```python
import glob
from lfp_extracters import * 
from scipy.signal import resample 
import matplotlib.pyplot as plt 
%matplotlib inline
```

## Load Data

```python
# get files 
A = glob.glob('data/data/Stimulus_A/*.csv')
B = glob.glob('data/data/Stimulus_B/*.csv')
files = list(A) + list(B)
len(files)
```

## Compute and Save LFP

```python
# for each file, load individually into DF (load h5) 
i = 0

for f in files: 
    print('Saving file ' + str(i+1) + ' of 1200')
    
    n1 = np.genfromtxt(f, delimiter=',')
    n1 = np.array(n1)
    n1 = n1[:32556] #EXACTLY ONE SECOND
    n1 = np.transpose(n1)
    
    t = n1[16] 
    
    # apply LFP 
    n2 = extract_lfp_bp(n1[0:16])
#     n2 = n2.append(t)
    
    # downsample to 600Hz 
    # Original sample is 32556/fs = 1 s in duration. New sample will be 6000/fs_new = 1s in duration.
    n3, newt = resample(n2, 6000, t = t, axis = 1)
    
    n4 = np.vstack((n3, newt))
    
    if 'A' in f: 
        stim = 'A'
    else: 
        stim = 'B'
    
    #and resave
    path = 'data/data/Stimulus_' + stim + '/LFP_15_300/'
    filename = 'filtered_stim_' + stim + '_' + str(i) + '.csv'
    np.savetxt(path + '/' + filename, n4, delimiter=',')
    
    i+=1
```

## Plot LFP and Downsampled Signal

```python
# make a plot
# load a single file 
n1 = np.genfromtxt('data/data/Stimulus_A/stim_A_2.csv', delimiter=',')
n1 = np.array(n1)
n2 = n1[:32556]  #EXACTLY ONE SECOND
n1 = np.transpose(n1)
n2 = np.transpose(n2)
# plot the whole thing 
plt.plot(n1[16] - n2[16][0], n1[0], n2[16] - n2[16][0], n2[0])
plt.show()
# plot LFP & plot downsampled LFP over it 
t = n2[16]
t = t - t[0]
n3 = extract_lfp_bp(n2[0:16])
n4, newt = resample(n3, 6000, t = t, axis = 1)
plt.plot(t[0:200], n3[0][0:200]) 
plt.scatter(newt[1:38], n4[0][1:38], c = 'r')
plt.show()

plt.plot(t, n3[0])
plt.show
```
